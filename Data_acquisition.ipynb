How many Purpleair sensors are there in the world?

```python
p = SensorList()  # Initialized 22,453 sensors!
p.to_dataframe(sensor_filter='outside',channel='parent')  #channel = parent or child, sensor_filter = 'outside', 'useful', 'family', and  'all'
#you can save this to a variable for further analysis
kk = p.to_dataframe(sensor_filter='outside',channel='parent') 
kk.to_excel('purpleair_sensors.xlsx')
kk
```

How is the air pollution levels in Hazelwood Canary?

```python
se = Sensor(14633)
df = se.parent.get_historical(2,thingspeak_field='primary')
df
```

Let's look at the sensors in West Africa

```python
kk_africa = kk[kk['lat']>3]
kk_africa = kk_africa[kk_africa['lat']<15]
kk_africa = kk_africa[kk_africa['lon']>3]
kk_africa = kk_africa[kk_africa['lon']<15]
kk_africa

hd = kk[(kk['lat'] > 3) & (kk['lon'] >3) & (kk['lat'] < 15)  & (kk['lon'] <15)]
hd
```

Can we extract the PM2.5 values from all the locations in West Africa?

```python
westafr_pm25 = []
loc = []
for i,ii in enumerate(hd.index):
  
  we = Sensor(ii)
  xc = we.parent.get_historical(1,thingspeak_field='primary')
  
  if len(xc) > 5000:
    westafr_pm25.append(xc['PM2.5 (CF=1) ug/m3'].values)
    print([i,ii,len(xc)])
    loc.append(ii)

westafr = pd.DataFrame(westafr_pm25).T
westafr.columns = loc
westafr
#westafr.plot()
```
