# API Access

Most IoT services have API for data access.  This is the recommended approach in most cases.

## However, ....

```python
import requests
my_api = '912D94F9-38F9-11ED-B5AA-42010A800006'
def getSensorData(sensor_index):
  my_url = 'https://api.purpleair.com/v1/sensors/'+str(sensor_index)
  my_headers = {'X-API-Key':my_api}
  r = requests.get(my_url,headers=my_headers)

  return r

def getSensorHistory(sensor_index):
  fields_list = ['pm2.5_atm_a', 'pm2.5_atm_b', 'pm2.5_cf_1_a', 'pm2.5_cf_1_b', 'humidity', 'humidity_b', 
               'temperature', 'temperature_b', 'pressure', 'pressure_b']
  my_url = 'https://api.purpleair.com/v1/sensors/57207/history'
  for i,f in enumerate(fields_list):
      if (i == 0):
          my_url += f'?fields={f}'
      else:
          my_url += f'%2C{f}'
  my_headers = {'X-API-Key':my_api}
  r = requests.get(my_url,headers=my_headers)

  return r
```
